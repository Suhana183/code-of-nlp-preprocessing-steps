```python
import re

text = "<p>Hello <b>World</b></p>"

clean_text = re.sub(r'<.*?>', '', text)

print(clean_text)
```

```python
text = "Hello!!! How are you?"

for p in "!,?.:":
    text = text.replace(p, "")

print(text)
```

```python
text = "u cant come"

text = text.replace("u", "you")
text = text.replace("cant", "cannot")

print(text)
```

```python
import re

text = "Visit https://google.com now"

clean_text = re.sub(r'http\S+', '', text)

print(clean_text)
```

```python
text="NLP IS FUN"
text=text.lower()
print(text)
```

```python
text="nlp is fun"
text=text.upper()
print(text)
```

```python
text = "I am happy ðŸ˜Š"
text = text.replace("ðŸ˜Š", " happy")

print(text)
```

```python
text = "I am learning NLP"
stop_words = ["I", "am"]

print([w for w in text.split() if w not in stop_words])
```

```python
text = "I love NLP"

tokens = text.split()

print(tokens)
```

```python
from nltk.stem import PorterStemmer

stem = PorterStemmer()

print(stem.stem("playing"))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer


nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

words = ["running", "playing", "cars", "better", "studies"]

for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt'); nltk.download('stopwords'); nltk.download('punkt_tab')
print([w for w in word_tokenize("This is stopword removal example") if w.lower() not in stopwords.words('english')])
```
